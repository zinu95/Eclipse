```python
import matplotlib.pyplot as plt
```

### 선 그래프

```python
y = [2,4,6]
plt.plot(y) # 그래프 그리기
plt.show()
```

```python
y = [2,4,6]
x = [1,2,3]
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,linestyle = '--') # (줄의 모양)
plt.show()
```

```python
# marker
plt.plot(y,marker='o')
plt.show()
```

```python
# line width (줄의 두께)
plt.plot(x,y,linewidth=5)
plt.show()
```

```python
# line color
plt.plot(x,y,color='r')
plt.show()
```

```python
x=[1,2,3,4,5,6,7]
y=[1,4,5,8,9,5,3]
plt.plot(x,y,linestyle='--',marker='o',color='r')
plt.show()
```

```python
x=[1,2,3,4,5,6,7]
y=[1,4,5,8,9,5,3]
plt.plot(x,y,linestyle=':',marker='D',markersize='10', color='b')
plt.show()
```

```python
# 범위 지정 (그림의 영역 지정)
# xlim : x 범위 지정
# ylim : y 범위 지정
x = [0,1,2,3,4,5,6]
y = [1,4,5,8,9,5,3]
plt.plot(x,y,ls = '--', lw = 5 , marker = 'o', ms = 15, mec = 'g', mew = 5, mfc = 'r')
plt.ylim(-4,12)
plt.xlim(-1,8)
plt.show()
```

```python
# tick 설정
# xticks : x축 틱 설정
# yticks : y축 틱 설정

x = [0,1,2,3,4,5,6]
y = [1,4,5,8,9,5,3]
plt.plot(x,y,ls = '--', lw = 5 , marker = 'o', ms = 15, mec = 'g', mew = 5, mfc = 'r')
plt.xticks([0,3,6])
plt.yticks([1,5,9])
plt.show()
```

```python
# grid
x = [0,1,2,3,4,5,6]
y = [1,4,5,8,9,5,3]
plt.plot(x,y,ls = '--', lw = 5 , marker = 'o', ms = 15, mec = 'g', mew = 5, mfc = 'r')
plt.xticks([0,3,6])
plt.yticks([1,5,9])
plt.grid()
plt.show()
```

```python
# 여러개 그래프 동시에 그리기
x = [1,2,3,4]
y = [2,4,6,8]
z = [3,6,9,12]
plt.plot(x, color='r')
plt.plot(y)
plt.plot(z)
plt.show()
```

```python
plt.plot(x, color='r') # 그래프 그리기
plt.show() # 위에 까지 그린 그래프를 출력해 준다.
plt.plot(y)
plt.show()
plt.plot(z)
plt.show()
```

```python
# 범례 표시
# plt.legend()
plt.plot(x, color='r', label='x')
plt.plot(y, label='y')
plt.plot(z, label='z')
plt.legend()
plt.show()
```

```python
# 차트는 한글지원을 안하므로 한글 폰트를 불러와야 함
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname='c:/Windows/Fonts/malgun.ttf').get_name()
rc('font', family=font_name)
```

```python
# label -> 각각의 축에 이름 붙여주기
# title -> 전체 그래프 이름
plt.plot(x, color='r', label='엑스')
plt.plot(y, label='와이')
plt.plot(z, label='제트')
plt.xlabel('x축')
plt.ylabel('y축',rotation=0) # ylabel의 기본 회전값 : 90도 -> rotation = 0도로 보기 쉽게 돌리기
plt.title('x,y,z 그래프')
plt.legend()
plt.show()
```

```python
data = pd.read_csv('Traffic_Accident_2017.csv', encoding='euc-kr')
data
```

```python
# 요읿별 가장 많은 사고가 발생했을지?
temp = data['요일'].value_counts()
temp
```

```python
temp[['월','화']]
```

```python
y = temp[['월','화','수','목','금','토','일']]
x = y.index
```

```python
y.index
```

```python
# bar chart, x축에 대해 값 지정 X
# 막대그래프
plt.bar(x,y)
plt.title('요일별 사고량')
plt.xlabel('요일')
plt.ylabel('사고건수', rotation=0)
plt.show()
```

```python
car_to_car = data[data['사고유형_대분류']=='차대차']
car_to_car
```

```python
# groupby() -> 같은 값을 하나로 묶어 통계 또는 집계 결과를 얻기 위해 사용하는 것.
car_to_car_place = car_to_car[['사상자수','발생지시도']].groupby('발생지시도').sum()
car_to_car_place
```

```python
car_to_car_place.index
```

```python
car_to_car_place.values
```

```python
x = car_to_car_place.index
```

```python
y = car_to_car_place.values
```

```python
plt.figure(figsize=(15,8)) # 그래프 전체 크기 조절
plt.plot(x,y)
plt.xlabel('도시')
plt.ylabel('사상자수', rotation=0)
plt.show()
```
