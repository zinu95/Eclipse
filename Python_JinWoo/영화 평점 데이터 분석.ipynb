```python
import numpy as np
```

```python
# np.loadtxt('파일명.확장자', 파일에서 사용한 구분 문자, 데이터 타입 지정)
# delimiter -> 구분 문자
data = np.loadtxt('ratings.txt', delimiter='::', dtype = 'int')
data
```

```python
# 데이터 구조 확인하기
# 크기, 요소의 개수, 차원
print('크기 :', data.shape)
print('요소의 개수 :', data.size)
print('차원 :', data.ndim)
```

### 1번 user 전체 정보 출력

```python
# 1. id 정보만 가져오기
data[:,0]
```

```python
# 2. id가 1인 데이터만 가져오기
#     ↑조건 (불리언인덱싱)
data[:,0] == 1
```

```python
# 3. 불리언인덱싱
data[data[:,0]==1]
```

### 전체 사용자의 평점 평균 구하기

```python
# 1. 평점 데이터만 인덱싱
data[:,2]
```

```python
# 2. 평점 데이터 평균값 -> mean()
np.mean(data[:,2])
```

### 1번 사용자의 평점 평균 구하기

```python
# 1번 + 2번 문제

# 1. 1번 사용자의 전체 데이터 구하기
user_1 = data[data[:,0]==1]
# 2. 1번 사용자의 전체 데이터에서 평점 가져오기
ratings_1 = user_1[:,2]
# 3. 가져온 평점에서 평균을 구하기
np.mean(ratings_1)
```

### 4. 각 사용자의 평점 평균 구하기

```python
# 데이터의 중복을 제거하고 출력
# np.unique
user_id = np.unique(data[:,0])
user_id
```

```python
data[:,0]
```

```python
# 사용자별 평점 평균을 구하는 for문 만들기
user_id = np.unique(data[:,0])
list1 = []
# 1. 반복문 사용
for user in user_id:
    # 2. id 정보 가져오기
    u_id = data[:,0]
    # 3. 사용자별 정보 가져오기 (불리언)
    #           ↓data[:,0]==1, 1 끝나면 data[:,0]==2, ... , data[:,0]==6040
    user_bol = u_id == user # 1
    # 4. 사용자별 정보에서(불리언 인덱싱) 평점만 가져오기
    ratings = data[user_bol][:,2]
    # 5. 평점에서 평균내고
    r_mean = np.mean(ratings)
    # 6. 평균낸 새로운 데이터를 리스트에 담아주기
    list1.append([user, r_mean])
    print('{}번 사용자의 평점 : {}'.format(user, r_mean))
```

```python
# 간단하게 코드를 줄여보자!
user_id = np.unique(data[:,0])
list3 = []

for user in user_id:
    user_mean = data[data[:,0]==user, 2].mean()
    list3.append([user, user_mean])
    # 리스트에 담기 [사용자 아이디, 사용자 아이디에 따른 평점 평균]
```

```python
list3
```

```python
# 파일을 외부로 내보낼 때 쓰는 함수
# np.savetxt('파일명.확장자', 해당 array 변수, 구분자)
np.savetxt('user_id_mean.csv',list3,delimiter=',')
# 이 때 저장되는 파일 경로는 해당 notebook이 있는 위치폴더
```
