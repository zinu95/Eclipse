### 조건별로 출력하기
- 불리언 인덱싱

- 면적이 130 이상인 아파트행들의 전체 열 출력
- 면적이 130 이상인 값 = True
- 면적이 130 미만인 값 = False

```python
df['면적'] >= 130
```

```python
# 인덱싱을 불리언 값으로 하는 것 = 불리언 인덱싱
# True의 값을 가진 행 출력
# False의 값을 가진 행 출력 X
df[df['면적'] >= 130]
```

```python
# 인덱서
df.loc[df['면적']>=130,:]
```

- 면적이 130 이상인 아파트의 가격만 출력
- 행 : 면적이 130 이상
- 열 : 가격

```python
df.loc[df['면적']>=130,'가격']
```

- 가격 컬럼은 1당 만원 즉, 1.5억 = 15000
- 면적이 130 이상이고 가격은 1.5억 미만인 아파트의 가격만 출력
- 행 : 면적이 130 이상,& 가격이 1.5억 미만 (즉, 두 개 다 True)
- 열 : 가격

```python
df.loc[(df['면적']>=130) & (df['가격']<15000), '가격']
```

- 면적이 130 이상이고 가격은 1.5억 미만인 자료의 가격의 내림차순 출력

```python
# sort_values(기준이 될 컬럼)
df.loc[(df['면적']>=130) & (df['가격']<15000), '가격'].sort_values(ascending=False)
```

###  원하는 값 바꾸기
- 가격 컬럼 값 바꾸기
- 가격 컬럼의 값이 800 이하면 모두 다 0으로 바꾸기

```python
# 가격이 800 이하인 행의 가격 컬럼 출력
df[df['가격']<=800]['가격'] = 0 # 구버전에선 찾기 어려운 방법 (편법)
```

```python
df.loc[df['가격']<=800,'가격'] = 0 # 정석적인 방법
```
